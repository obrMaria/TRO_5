В книжном магазине проводится анализ по продажам.

Имеются данные по кол-ву проданных эпических, лирических и драматических произведений.

Нужно сравнить их продажи за последний месяц и определить, какой жанр литературы больше нравится покупателям



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# данные по продажам книг за месяц
genres = ['Эпические', 'Лирические', 'Драматические']
sales = [300, 150, 200]
```

```python
explode = (0.1, 0.1, 0.15)

fig, ax = plt.subplots()

ax.pie(sales, labels=genres, autopct='%1.1f%%', shadow=True, explode=explode,
wedgeprops={'lw':1, 'ls':'--','edgecolor':"k"}, rotatelabels=True)

ax.axis("equal")
```

Данная круговая диаграмма показывает процент покупок книг кадого из жанров. Как видно из диаграммы самый привлекаемый жанр - это Эпические (46,2%), затем уже Лирические и Драматические, процент которых составляет 23,1% и 30,8%, соответсвенно
